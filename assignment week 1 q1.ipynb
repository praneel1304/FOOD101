```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('ab.jpg')  
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


def sharpenedimage(img):
        img = np.array(img, dtype=float)
        kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
        h, w= np.shape(img)
        output = np.zeros_like(img)
        for i in range(1, h-1):
            for j in range(1, w-1):
                
                    
                   r = img[i-1:i+2, j-1:j+2]
                   output[i, j] = np.sum(r * kernel)
        
        return output.astype(np.uint8)
def blurimage(img):
        img = np.array(img, dtype=float)
        kernel = np.array([[1, 2, 1],
                          [2, 4, 2],
                          [1, 2, 1]]) / 16
       
        h,w = np.shape(img)
        output = np.zeros_like(img)

        for i in range(1, h-1):
            for j in range(1, w-1):
                
                    
                   r = img[i-1:i+2, j-1:j+2]
                   output[i, j] = np.sum(r * kernel)

        
        return output.astype(np.uint8)

def edge_detec(img):
        img = np.array(img, dtype=int)
        kernel = np.array([[-1, -1, -1],
                          [-1, 8, -1],
                          [-1, -1, -1]]) 
        h, w = np.shape(img)
        output = np.zeros_like(img)

        for i in range(1, h-1):
            for j in range(1, w-1):
               
                    
                   r = img[i-1:i+2, j-1:j+2,]
                   output[i, j] = np.sum(r * kernel)

        
        return output.astype(np.uint8)

              


```

```python
plt.imshow(img,cmap = 'gray')
```

```python
##blur image
plt.imshow(blurimage(img),cmap = "gray")
```

```python
##for sharpened image

plt.imshow(sharpenedimage(blurimage(img)),cmap = "gray")
```

```python
plt.imshow(edge_detec(blurimage(img)),cmap = 'gray')
```
